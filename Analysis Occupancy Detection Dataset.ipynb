```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ganjar87/data_science_practice/main/datatraining.txt')
data
```

#Probability Mass Function

```python
# Probability Mass Function

from scipy.stats import norm

sns.histplot(data=data, x='Occupancy',stat="probability")      
plt.xticks(rotation='45')                                  
plt.show()
```

```python
data['Occupancy'].value_counts(normalize = True)
```

#Analisis Distribusi

```python
# DataFrame untuk Occupancy = 0
occupancy_0 = data[data['Occupancy']==0]
occupancy_0.head()
```

```python
# DataFrame untuk Occupancy = 1
occupancy_1 = data[data['Occupancy']==1]
occupancy_1.head()
```

```python
# Distribusi Temperature saat Occupancy = 0
min = occupancy_0['Temperature'].min()
max = occupancy_0['Temperature'].max()

mean = occupancy_0['Temperature'].mean()
std = occupancy_0['Temperature'].std()


x_pdf = np.linspace(min, max)
y_pdf = norm.pdf(x_pdf, mean, std)

sns.histplot(data=occupancy_0, x='Temperature',stat="density")
plt.plot(x_pdf, y_pdf, 'r', lw=2, label='pdf')       
plt.legend()                                            
plt.show()
```

```python
# Distribusi Temperature saat Occupancy = 1
min = occupancy_1['Temperature'].min()
max = occupancy_1['Temperature'].max()

mean = occupancy_1['Temperature'].mean()
std = occupancy_1['Temperature'].std()


x_pdf = np.linspace(min, max)
y_pdf = norm.pdf(x_pdf, mean, std)

sns.histplot(data=occupancy_1, x='Temperature',stat="density")
plt.plot(x_pdf, y_pdf, 'r', lw=2, label='pdf')       
plt.legend()                                            
plt.show()
```

```python
sns.histplot(data=data, x='Temperature', hue='Occupancy')     
plt.xticks(rotation='45')                                  
plt.show()
```

Penjelasan : terdapat perbedaan Distribusi Temperature saat Occupancy = 1 dan saat Occupancy = 0. Dapat dilihat pada histogram di atas, saat Occupancy = 1 Suhu cenderung meningkat dikisaran 21-23 derajat. Sedangkan saat Occupancy = 0 suhu lebih banyak berada dikisaran 19-21 derajat. Hal ini menunjukkan bahwa saat tidak berpenghuni, suhu akan menjadi lebih dingin. Akan tetapi, Keduanya memiliki kemiripan pada garis Probability Density Function yang ditampilkan.

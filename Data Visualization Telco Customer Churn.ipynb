```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv ('https://raw.githubusercontent.com/ganjar87/data_science_practice/main/WA_Fn-UseC_-Telco-Customer-Churn.csv')
dataset
```

#Pie Chart

```python
pie_chart = dataset['InternetService'].value_counts()
pie_chart.plot.pie(autopct='%.2f%%')

plt.show()
```

#Bar Plot

```python
existing_data = dataset[(dataset.Churn == 'No')]
churn_data = dataset[(dataset.Churn == 'Yes')]

data1 = existing_data['PaymentMethod'].value_counts()
data2 = churn_data['PaymentMethod'].value_counts()

df = pd.concat([data1, data2], keys=['existing','churn'], axis=1)
df.plot(kind='bar')

plt.xticks(rotation=45)
plt.title('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
#Monthly Charges Histogram
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
existing_data['MonthlyCharges'].plot.hist(label='Existing')
plt.ylabel('Count')
plt.xlabel('Monthly Charges')
plt.ylim([0,1500])
dataan1 = existing_data['MonthlyCharges']
plt.text(60, 1100, "Std dev existing: " + str(round(dataan1.std(),2)), 
        style = 'italic', fontsize=10)
plt.text(60, 1200, "Mean existing: " + str(round(dataan1.mean(),2)), 
        style = 'italic', fontsize=10)

churn_data['MonthlyCharges'].plot.hist(label='Churn')
plt.ylabel('Count')
plt.xlabel('Monthly Charges')
plt.ylim([0,1500])
dataan2 = churn_data['MonthlyCharges']
plt.text(60, 900, "Std dev churn: " + str(round(dataan2.std(),2)), 
        style = 'italic', fontsize=10)
plt.text(60, 1000, "Mean churn: " + str(round(dataan2.mean(),2)), 
        style = 'italic', fontsize=10)
plt.title('Monthly Charges')
plt.legend()

#Tenure Histogram
plt.subplot(1, 2, 2)
existing_data['tenure'].plot.hist(label='Existing')
plt.ylabel('Count')
plt.xlabel('Tenure')
plt.ylim([0,1100])
dataan3 = existing_data['tenure']
plt.text(20, 690, "Std dev existing: " + str(round(dataan3.std(),2)), 
        style = 'italic', fontsize=10)
plt.text(20, 750, "Mean existing: " + str(round(dataan3.mean(),2)), 
        style = 'italic', fontsize=10)

churn_data['tenure'].plot.hist(label='Churn')
plt.ylabel('Count')
plt.xlabel('Tenure')
plt.ylim([0,1100])
dataan4 = churn_data['tenure']
plt.text(20, 570, "Std dev churn: " + str(round(dataan4.std(),2)), 
        style = 'italic', fontsize=10)
plt.text(20, 630, "Mean churn: " + str(round(dataan4.mean(),2)), 
        style = 'italic', fontsize=10)
plt.title('Tenure')


plt.legend()
plt.show()
```

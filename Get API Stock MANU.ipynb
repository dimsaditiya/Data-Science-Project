```python
import pandas as pd
import requests
```

#Pengambilan Data

```python
# Saham Manchester United (MANU) dalam Monthly
link = 'https://www.alphavantage.co/query?'

queries = {
    'apikey': 'SUST109DVHXHS5XS',
    'symbol': 'MANU',
    'function': 'TIME_SERIES_MONTHLY'
}

response = requests.get(link, params=queries)
print(response.status_code)
```

```python
data = response.json()
data
```

##Processing Data

```python
symbol = data['Meta Data']['2. Symbol']
days = data['Monthly Time Series']
date = list(days.keys())
```

```python
processed_data = []
for i in range(len(date)):
    tanggal = date[i]
    open = days[date[i]]['1. open']
    high = days[date[i]]['2. high']
    low = days[date[i]]['3. low']
    close = days[date[i]]['4. close']
    volume = days[date[i]]['5. volume']
    row = [symbol, tanggal, open, high, low, close, volume]
    processed_data.append(row)
```

```python
processed_data = pd.DataFrame(processed_data, columns=['symbol', 'date', 'open', 'high', 'low', 'close', 'volume'])
processed_data.head(14)
```
